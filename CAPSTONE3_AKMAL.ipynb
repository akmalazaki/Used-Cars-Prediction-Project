### **1.5 Metric Evaluation**

Dalam proses pembersihan data, kami tidak menghapus seluruh data yang memiliki outlier. Oleh karena itu, data tetap akan memiliki beberapa outliers. Karena kondisi ini, model ini akan menggunakan metrik evaluasi yang tidak sensitif terhadap outliers. Metrik seperti Mean Squared Error (MSE), Root Mean Squared Error (RMSE), dan Root Mean Squared Percentage Error (RMSPE) tidak cocok digunakan pada dataset ini karena sensitif terhadap data outliers metrik-metrik tersebut juga kurang cocok digunakan pada variabel target yang memiliki ragam dan rentang nilai yang sangat besar seperti mata uang.

Sebagai gantinya, metrik evaluasi yang akan digunakan adalah metrik-metrik yang tahan dan mampu menghandle outlier dengan baik. Adapun matrix evaluation yang kami gunakan untuk mengukur kinerja model adalah sebagai berikut:

1. `Mean Absolute Error (MAE)`: MAE mengukur rata-rata dari selisih absolut antara nilai aktual dan nilai yang diprediksi oleh model. Nilai MAE yang lebih rendah menunjukkan kinerja yang lebih baik. Alasan kami menggunakan MAE adalah karena variabel target dalam hal ini adalah variabel Price memiliki outlier sehingga kami memutuskan untuk menggunakan MAE dan MAE merupakan metrics yang kurang sensitif terhadap pencilan/outlier. MAE tahan terhadap outlier karena mengambil nilai mutlak dari selisih antara nilai aktual dan nilai yang diprediksi, sehingga outlier dengan kesalahan besar hanya memberikan kontribusi yang setara tanpa memberikan "bobot" ekstra pada kesalahan tersebut.
2. `Mean Absolute Percentage Error (MAPE)`: MAPE mengukur rata-rata dari selisih persentase antara nilai aktual dan nilai yang diprediksi oleh model. Nilai MAPE yang lebih rendah menunjukkan kinerja yang lebih baik. Alasan kami menggunakan MAPE sama seperti MAE yaitu karena pada variabel target (Price) memiliki outlier dan MAPE kurang sensitif terhadap outlier. MAPE tahan terhadap outlier karena mengukur kesalahan relatif sebagai persentase, dan oleh karena itu, outlier dengan kesalahan besar tidak memiliki dampak yang sangat besar pada metrik ini. MAPE mempertimbangkan perbandingan relatif antara nilai aktual dan nilai yang diprediksi, sehingga outlier dengan nilai besar tidak mengubah proporsi kesalahan yang signifikan. Ini membuat MAPE lebih tahan terhadap dampak outlier yang signifikan pada hasil evaluasi.
3. `Root Mean Squared Logarithmic Error (RMSLE)`: RMSLE adalah metrik yang mengukur perbedaan antara logaritma alami dari nilai aktual dan logaritma alami dari nilai yang diprediksi oleh model. Alasan kami mempertimbangkan matrix ini adalah karena RMSLE juga cenderung lebih tahan terhadap outlier. Matrix ini tahan terhadap outlier karena perbedaan persentase antara nilai aktual dan nilai prediksi diukur setelah transformasi logaritma, sehingga outlier yang mungkin memiliki dampak besar pada metrik lain menjadi lebih terkendali. Namun karena mengambil logaritma data, hasil RMSLE cenderung lebih sulit untuk diinterpretasikan secara langsung dalam konteks asli data, sehingga RMSLE hanya kami gunakan sebagai pertimbangan dalam memilih model terbaik pada benchmarking model.
4. `R-squared (R2)`: R2 mengukur seberapa besar variasi nilai Y yang dapat dijelaskan oleh model. Nilai yang lebih tinggi menunjukkan kinerja yang lebih baik. Sebenarnya R2 kurang cocok digunakan dalam model ini, karena terdapat data outlier pada variabel target dan variabel target memiliki ragam dan rentang nilai yang sangat besar, sehingga matrix ini hanya kami gunakan sebagai pertimbangan dalam memilih model terbaik pada benchmarking model.

### **4.5 Handling Extreme Outliers**


Sebelum kita melakukan filtering pada data, penting untuk mendeteksi outlier terlebih dahulu. Outlier dapat memiliki dampak yang signifikan pada integritas data kita. Mereka dapat meningkatkan variabilitas kesalahan dan mengurangi kekuatan dari uji statistik. Outlier juga dapat mempengaruhi estimasi dan asumsi dasar dalam model statistik, seperti regresi.

Namun, problemnya adalah bagaimana cara kita menghandle outlier? Apakah ada outlier positif (nilai yang sangat tinggi) dan outlier negatif (nilai yang sangat rendah) dalam dataset kita? Bagaimana kita dapat menghapus outlier dan membersihkan dataset tanpa menghilangkan terlalu banyak informasi yang mungkin berguna?

Semua pertanyaan ini akan dijawab dalam proyek kita di bawah ini. Dalam langkah-langkah berikutnya, kita akan mendiskusikan metode dan teknik untuk mendeteksi dan mengatasi outlier dalam dataset.

Mari kita membuat fungsi untuk mengecek seluruh data yang memiliki data *outlier* pada kolom numerik

```python
# Membuat fungsi untuk mengecek outliers
def find_outlier(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')  
```

### **5.4 Multicollinearity**

Multikolinearitas menjadi masalah dalam analisis regresi ketika dua atau lebih variabel independen dalam model regresi berkorelasi kuat satu sama lain. Tingkat multikolinearitas dapat diukur dengan mengamati statistik seperti Variance Inflation Factor (VIF) atau dengan memeriksa matriks korelasi antara variabel-variabel independen.
* Multikolinearitas dianggap tinggi ketika VIF melebihi 10
* Multikolinearitas sedang biasanya terjadi ketika VIF berada dalam kisaran 5 hingga 10
* Multikolinearitas dianggap rendah ketika VIF kurang dari 5

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

data = df_clean.select_dtypes(include=['number'])
data = data.drop(['Price'], axis=1)

def calc_vif(X):
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(data.shape[1])]
    return(vif)
```

```python
calc_vif(data)
```

### **6.3.1 Choose Best Model**

Tahap 1:

Pada tahap ini, kita akan mencari model terbaik dengan menggunakan model-model yang telah disebutkan dalam bagian "Modeling". Setiap model akan diujicoba pada data pelatihan, kemudian kita akan mengamati hasil metrik evaluasinya. Selain itu, kita juga akan menggunakan fungsi tambahan yaitu TransformedTargetRegressor. Fungsi ini digunakan untuk membuat skala pada variabel *Target* menjadi lebih kecil sehingga distribusi datanya lebih mendekati distribusi normal. Setelah *Target* diproses, fungsi ini akan mengembalikan nilai target ke skala aslinya, sehingga hasil evaluasi model dapat dioptimalkan dengan lebih baik.

Pengujian pada data training

```python
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state = 100)
rf = RandomForestRegressor(random_state = 100)
gdt = GradientBoostingRegressor(random_state = 100)
xgb = XGBRegressor(random_state = 100)

ttr_lr = TransformedTargetRegressor(lr, func = np.log, inverse_func = np.exp)
ttr_knn = TransformedTargetRegressor(knn, func = np.log, inverse_func = np.exp)
ttr_dt = TransformedTargetRegressor(dt, func = np.log, inverse_func = np.exp)
ttr_rf = TransformedTargetRegressor(rf, func = np.log, inverse_func = np.exp)
ttr_gdt = TransformedTargetRegressor(gdt, func = np.log, inverse_func = np.exp)
ttr_xgb = TransformedTargetRegressor(xgb, func = np.log, inverse_func = np.exp)

models = [ttr_lr, ttr_knn, ttr_dt, ttr_rf, ttr_gdt, ttr_xgb]

score_mae = []
score_mape = []
score_rmsle = []
score_r2 = []

for model in models:
    
    cross_val = KFold(n_splits = 5, shuffle = True, random_state = 100)
    estimator = Pipeline([
        ('preprocessing', transformer),
        ('model', model)
    ])

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'neg_mean_absolute_error', 
        error_score = 'raise'
        )

    score_mae.append(model_cv_mae.mean())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'neg_mean_absolute_percentage_error', 
        error_score = 'raise'
        )

    score_mape.append(model_cv_mape.mean())
    
       # RMSLE
    model_cv_rmsle = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'neg_mean_squared_log_error', 
        error_score = 'raise'
        )

    score_rmsle.append(model_cv_rmsle.mean())

       # R2
    model_cv_r2  =  cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'r2', 
        error_score = 'raise'
        )

    score_r2.append(model_cv_r2.mean())

score_rmsle = np.sqrt(abs(np.array(score_rmsle)))
```

Berikut merupakan hasil dari matrix evaluation pada setiap model:

```python
Model_name = ['Linear Regression', 'KNN Regression', 'Decision Tree Regression', 'Random Forest Regression', 'Gradient Boost Regression','Extreme Gradient Boost']

train_score = pd.DataFrame({'Model': Model_name,
                            'Score MAE': score_mae,
                            'Score MAPE': score_mape,
                            'Score RMSLE': score_rmsle,
                            'Score R2': score_r2
                            })

train_score
```

Dari tabel di atas, terlihat bahwa model Extreme Gradient Boost merupakan model terbaik ketika diujicobakan pada data pelatihan. Hasil evaluasi metrik dari model Extreme Gradient Boost menunjukkan bahwa setiap metrik memiliki nilai yang paling rendah dibandingkan dengan model lainnya dengan nilai MAE sebesar 11541.906594, MAPE sebesar 0.201056, dan RMSLE sebesar 0.275380. Selain itu, XGBoost juga memiliki nilai R-Square yang terbesar, menunjukkan bahwa model Extreme Gradient Boost mampu merepresentasikan sekitar 78% variabilitas dalam data target.

Berdasarkan hasil tersebut, model yang akan digunakan adalah Extreme Gradient Boost, karena hasil metrik evaluasinya yang paling baik dibandingkan dengan model lainnya.

Selanjutnya mari kita bahas konsep tentang model Extreme Gradient Boost. XGBoost adalah metode ensemble learning yang termasuk dalam kategori "boosting." Dalam ensemble learning, beberapa model lemah digabungkan untuk membuat model yang lebih kuat. Boosting adalah jenis ensemble learning di mana model-model lemah ditambahkan secara berurutan, dan setiap model berusaha untuk memperbaiki kesalahan model sebelumnya.

Konsep dasar dari XGBoost adalah sebagai berikut:

1. **Boosting**: Dalam boosting, model-model lemah (dalam konteks XGBoost, pohon keputusan) dibangun secara berurutan. Setiap model berusaha untuk memperbaiki kesalahan yang telah dibuat oleh model sebelumnya. Dengan demikian, setiap model "mengajar" model berikutnya cara memprediksi dengan lebih baik.

2. **Pembobotan Data**: Ketika data salah diprediksi oleh model, data ini diberi bobot tambahan. Bobot ini membuat model selanjutnya lebih memperhatikan data yang salah tersebut. Dengan demikian, model fokus pada kasus-kasus yang sulit atau salah diprediksi sebelumnya, yang dapat meningkatkan akurasi keseluruhan.

3. **Iterasi**: Setelah memberikan bobot tambahan pada data yang salah diprediksi, langkah selanjutnya adalah melakukan iterasi. Sejumlah model lemah (pohon keputusan dalam konteks XGBoost) dibangun secara berurutan, di mana setiap model berusaha untuk memperbaiki kesalahan model sebelumnya. Jumlah iterasi ini ditentukan oleh parameter n_estimators dalam XGBoost.

4. **Pembobotan Model**: Konsep pembobotan dijelaskan melalui parameter learning_rate dalam XGBoost. Learning rate mengontrol sejauh mana model berubah setiap kali ada iterasi tambahan. Dengan nilai learning rate yang lebih rendah, pembobotan data yang salah prediksi akan lebih halus dan perubahan model akan lebih kecil setiap iterasi, yang dapat membantu mencegah overfitting.

XGBoost adalah model machine learning yang kompleks, namun memiliki beberapa keunggulan, antara lain kemampuan untuk bekerja secara paralel membuatnya cocok untuk data berukuran besar. Hal ini memungkinkan penggunaan sumber daya komputer yang lebih efisien dan mempercepat proses pelatihan model. XGBoost juga mampu mengidentifikasi data yang salah diprediksi melalui pembobotan, namun XGBoost sulit diinterpretasikan, meskipun memiliki fitur-fitur interpretasi seperti peringkat fitur, XGBoost seringkali kompleks sehingga sulit untuk diinterpretasikan dengan baik. Ini dapat membuatnya kurang cocok untuk kasus di mana interpretabilitas model sangat penting. Penggunaan XGBoost dapat membebani sumber daya komputer, sehingga perlu mempertimbangkan kapasitas komputer yang tersedia.

Secara keseluruhan, XGBoost menggabungkan metode boosting dengan pembobotan data dan model, serta pengendalian perubahan model melalui learning rate. Kombinasi dari teknik-teknik ini menjadikan XGBoost sebagai salah satu pilihan yang kuat dan seringkali unggul dalam hal performa ketika digunakan untuk berbagai tugas machine learning.

Selanjutnya kita akan melakukan pengujian model pada data testing menggunakan model yang terpilih (Extreme Gradient Boost).

```python
ttr_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Membuat pipeline dengan transformer dan model (ttr_xgb)
model = Pipeline([
    ('preprocessing', transformer),
    ('model', ttr_xgb) 
])

# Melatih model dengan pipeline
model.fit(X_train, y_train)

# Melakukan prediksi pada data pengujian
y_pred = model.predict(X_test)

# Mengukur kinerja model pada data pengujian
score_mae_test = mean_absolute_error(y_test, y_pred)
score_mape_test = mean_absolute_percentage_error(y_test, y_pred)
score_rmsle_test = np.sqrt(mean_squared_log_error(y_test, y_pred))
score_r2_test = r2_score(y_test, y_pred)

# Membuat DataFrame dengan hasil evaluasi
score_before_tuning = pd.DataFrame({'Score MAE': [score_mae_test], 
                                   'Score MAPE': [score_mape_test], 
                                   'Score RMSLE': [score_rmsle_test],  
                                   'Score R2': [score_r2_test]}, 
                                  index=['Extreme Gradient Boost Test'])

score_before_tuning
```

### **6.3.2 Hyperparameter Tunning**

Tahap 2:

Pada tahap ini, kita akan melakukan hyperparameter tuning untuk mengidentifikasi parameter yang optimal pada model terpilih, yaitu Extreme Gradient Boost. Model Extreme Gradient Boost memiliki banyak parameter yang dapat diatur, namun kita perlu berhati-hati agar model tidak mengalami overfitting. Oleh karena itu, kita hanya memfokuskan perubahan pada empat parameter, yaitu:
1. `max_depth` (kedalaman pohon): mengontrol kedalaman maksimum pohon keputusan. Nilai yang lebih besar bisa meningkatkan kapasitas model, tetapi berpotensi menyebabkan overfitting.
2. `learning_rate` (tingkat pembelajaran): menentukan sejauh mana model berubah setiap iterasi. Learning rate yang rendah membuat perubahan lebih halus, membantu mencegah overfitting, tetapi memerlukan lebih banyak iterasi.
3. `n_estimators` (jumlah iterasi yang diinginkan): mengontrol berapa banyak pohon keputusan yang dibangun secara berurutan. Lebih banyak iterasi bisa meningkatkan kinerja. 
4. `Subsample`: hyperparameter ini mengontrol fraksi data yang akan digunakan pada setiap iterasi. Menguranginya dapat mengurangi overfitting. Dengan memilih parameter-parameter ini secara bijak, kita berharap dapat meningkatkan performa model tanpa mengorbankan kestabilan dan menghindari overfitting.

```python
max_depth = list(np.arange(7, 11, 1))
learning_rate = [0.01, 0.05, 0.1, 0.5, 0.9]
n_estimators = list(np.arange(75, 301, 25))
subsample = [0.8, 0.9, 1.0]

hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample
}
```

```python
xgb = XGBRegressor(random_state = 100, verbosity = 0)

estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

gs = GridSearchCV(
    estimator_xgb, 
    param_grid = hyperparam_space_xgb,
    cv = crossval, 
    scoring = ['neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_mean_absolute_error',
    verbose=10
)
```

```python
gs.fit(X_train, y_train)
```

```python
print('Extreme Gradient Boost')
print('Best_score:', gs.best_score_)
print('Best_params:', gs.best_params_)
```

Dari hasil *hyperparameter tuning* didapatkan hasil perubahan sebagai berikut:
1. learning_rate : 0.05
2. max_depth : 7
3. n_estimators : 300
4. Subsample : 0.8

Kemudian kita akan masukkan perubahan parameter tersebut ke model dan mencoba lagi ke data *testing*.

```python
xgb_tuning = TransformedTargetRegressor(gs.best_estimator_, func=np.log, inverse_func=np.exp)

xgb_tuning.fit(X_train, y_train)

y_pred_xgb_tuning = xgb_tuning.predict(X_test)

mae_xgb_tuning = mean_absolute_error(y_test, y_pred_xgb_tuning)
mape_xgb_tuning = mean_absolute_percentage_error(y_test, y_pred_xgb_tuning)
rmsle_xgb_tuning = np.sqrt(mean_squared_log_error(y_test, y_pred_xgb_tuning)) 

score_after_tuning = pd.DataFrame({'Score MAE': [mae_xgb_tuning],
                                   'Score MAPE': [mape_xgb_tuning],
                                   'Score RMSLE': [rmsle_xgb_tuning]},  
                                   index=["Extreme Gradient Boost Tuning"])
print("Nilai hasil tuning:")
score_after_tuning
```

Berikut merupakan hasil dari pengujian pada data testing sebelum dan sesudah hyperparameter tuning

```python
pd.concat([score_before_tuning, score_after_tuning]).drop(columns = 'Score R2')
```

Dari tabel di atas, hasil proses tuning menunjukkan bahwa model dapat memprediksi dengan lebih baik, sebagaimana ditunjukkan oleh penurunan nilai metrik evaluasi (nilai yang semakin rendah) pada MAE, MAPE dan juga RMSLE.
